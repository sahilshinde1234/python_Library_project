# Total Matches Played by a team

```python
#Concate : It is used to join two dataframe
#value_counts()  = it gives a count of unique values of the specific Feature
# reset_index = original indexing 
#The set_axis() method allows you set the index of the specified axis.

#reset_index = original indexing 
match_data=pd.concat([ipl_matches_data2['Team1'],ipl_matches_data2['Team2']]).value_counts().reset_index()
match_data.set_axis(['Teams','Total_Matches'],axis = 1 ,inplace = True)
match_data
```

 5)Mumbai indians have been played Maximum Matches in 15 Season of ipl Although,
Kochi Tuskers Kerala have been played Minimum Matches.

# Winning Ratio Analysis

```python
#round(30/85,2)*100  =35.0
win_data=match_data
win_data['Total_Wins']=ipl_matches_data2['WinningTeam'].value_counts().reset_index()['WinningTeam']
win_data['win_ratio']= round((win_data['Total_Wins']/win_data['Total_Matches']),2)*100
win_data
```

```python
plt.figure(figsize=(15,6.5))
sns.set_style('darkgrid')
wi = sns.barplot(data=win_data, x='win_ratio', y='Teams', ci=False, palette='viridis_r')
wi.set_xticklabels(win_data['win_ratio'], rotation=55, fontdict={'fontsize':10})
plt.show()
```

6)we can see that Gujrat Titans has best winning percentage but Gujrat Titans has low number of matches played.
So, overall Mumbai Indians has good winning percentage.

# Top 10 Batters that gets Player of Match Title

```python
#The set_axis() method allows you set the index of the specified axis.
player_of_match_data = ipl_matches_data2['Player_of_Match'].value_counts().reset_index()
player_of_match_data.set_axis(['Player_of_Match','Count'],axis = 1 ,inplace = True)
top_10_Player_of_match=player_of_match_data.head(10)
```

```python
plt.figure(figsize=(9,4))
plt.plot(top_10_Player_of_match['Count'], top_10_Player_of_match['Player_of_Match'], 'o-g')

plt.xlabel('Count')
plt.ylabel('Player of Match')

plt.title('Top 10 Player of Match')
plt.show()
```

7)Above analysis show that, with 25 man of the match trophies, AB de Villiers has most man of the match in IPL. 
While CH Gayle is in second number who has most man of the match in IPL.

# Top 10 IPL hosted cities

 We will analyze the top cities that hosted IPL most time.

```python
citraw = ipl_matches_data2['City'].value_counts().reset_index()
citraw.set_axis(['City','Hosted_Number'],axis = 'columns' ,inplace = True)
top10_city_host = citraw.head(10)
top10_city_host
```

```python
plt.figure(figsize=(16,8))
palette_color = sns.color_palette("ch:s=.25,rot=-.25")
sns.barplot(x='Hosted_Number', y='City',palette = palette_color, data=top10_city_host);
plt.title('Top 10 IPL hosted cities');
```

8)Mumbai hosted maximum matches of Indian Premium League.

# Number of Matches hosted in Seasons

```python
ipl_matches_data2["Season"]=pd.DatetimeIndex(ipl_matches_data2["Date"]).year
matches_in_season=ipl_matches_data2.groupby(['Season'])['ID'].count().reset_index().rename(columns={'ID':'matches'})
matches_in_season
```

```python
plt.figure(figsize=(12,5))

sns.barplot(x='Season', y='matches', data=matches_in_season);
plt.title('No of total matches per season');
```

 9)There is chart that show total number of matches held in each Season of IPL.
